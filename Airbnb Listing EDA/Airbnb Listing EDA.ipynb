```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
file_p = "listings.csv"
airbnb_data = pd.read_csv(file_p)
airbnb_data.head(20)
```

```python
# 1. Data Cleaning and Preparation
airbnb_data.dropna(inplace=True) 
```

```python
# Visualizing price distribution
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
print("Insight:")
print("The histogram shows that the majority of Airbnb listings have prices concentrated within a certain range, with a peak around $100-$200 per night.")
```

```python
# Visualizing availability distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='availability_365', data=airbnb_data)
plt.title('Availability Distribution')
plt.xlabel('Availability')
plt.ylabel('Count')
plt.show()
print("Insight:")
print("The count plot indicates that most listings have high availability, suggesting that many properties are frequently available for booking.")
```

```python
# Visualizing location distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='name', data=airbnb_data)
plt.title('Location Distribution')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
print("Insight:")
print("From the count plot, it appears that certain neighborhoods have higher numbers of Airbnb listings compared to others, indicating varying levels of demand and popularity across different areas.")
```

```python
# Visualizing property type distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='room_type', data=airbnb_data)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
print("Insight:")
print("The room plot illustrates the distribution of property types listed on Airbnb, with apartments being the most common type, followed by houses and condominiums.")
```
